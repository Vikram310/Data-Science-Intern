## QUESTION1 : WHO WILL WIN THE MATCH?

Performing EDA to find who has upper hand in winning the match

```python
rckr = matches.loc[((matches['team1'] == 'Royal Challengers Bangalore') | (matches['team1'] == 'Kolkata Knight Riders')) & 
              ((matches['team2'] == 'Kolkata Knight Riders') | (matches['team2'] == 'Royal Challengers Bangalore'))]
rckr
```

```python
rcb_vs_kkr = pd.DataFrame({'team1':rckr['team1'],'team2':rckr['team2'],'winner':rckr['winner'],'win_by_runs':rckr['win_by_runs'],'win_by_wickets':rckr['win_by_wickets']})
rcb_vs_kkr
```

```python
rcb_vs_kkr['winner'].value_counts()
```

```python
plt.figure(figsize=(8,4))
ax = sns.countplot(x='winner',data=rcb_vs_kkr, order = rcb_vs_kkr['winner'].value_counts().index)

plt.title('visualization based on number of wins RCV VS KKR')
plt.show()
```

```python
matches['winner'].value_counts()
```

```python
plt.figure(figsize=(8,4))
ax = sns.countplot(x='winner',data=matches, order = matches['winner'].value_counts().index)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.title('visualization based on number of wins')
plt.show()
```

From the above analysis KKR has more wins than RCB.
IN 24 matches played betwen KKR and RCB, KKR won 14 times and RCB won 10 times. So we can say that KKR has more chance of winning this match.

## QUESTION 2 : WHAT WILL TOSS WINNING TEAM CHOOSE?

```python
a = matches.loc[(matches['toss_winner'] == 'Royal Challengers Bangalore')]
b = matches.loc[(matches['toss_winner'] == 'Kolkata Knight Riders')]

#print(a.head())
#print(b.head())
#print(c.head())
#print(d.head())

a.head()
```

```python
a1 = pd.DataFrame({'toss_winner' : a['toss_winner'], 'toss_decision' : a['toss_decision']})
b1 = pd.DataFrame({'toss_winner' : b['toss_winner'], 'toss_decision' : b['toss_decision']})
```

```python
frames = [a1,b1]

toss = pd.concat(frames)
```

```python
toss
```

```python
toss['toss_winner'].unique()
```

```python
toss['toss_winner'].value_counts()
```

```python
plt.figure(figsize=(8,4))
ax = sns.countplot(x='toss_winner',data=toss, order = toss['toss_winner'].value_counts().index)

ax.set_xticklabels(ax.get_xticklabels())
plt.title('visualization based on number of toss wins')
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax = sns.countplot(x='toss_decision',data=toss, order = toss['toss_decision'].value_counts().index)

ax.set_xticklabels(ax.get_xticklabels())
plt.title('visualization based on number of toss wins')
plt.show()
```

```python
toss['toss_decision'].value_counts()
```

```python
w = pd.get_dummies(toss['toss_winner'],drop_first = True)
```

```python
dum = pd.concat([w,toss],axis=1)
dum
```

```python
dum.drop('toss_winner',axis = 1,inplace= True)
```

```python
dum
```

```python
z = pd.get_dummies(toss['toss_decision'],drop_first = True)
```

```python
rege = pd.concat([dum,z],axis=1)
rege
```

```python
rege.drop('toss_decision',axis = 1,inplace= True)
rege
```

USING LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
x=rege['Royal Challengers Bangalore']
y=rege['field']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 45)
```

```python
log = LogisticRegression()
log.fit(np.array(x_train).reshape(-1, 1),y_train)
```

```python
y_hat = log.predict(np.array(x_train).reshape(-1, 1)) #predictions are stored in y^(y_train should match y_hat)
```

```python
log.score(np.array(y_hat).reshape(-1, 1),y_train)
```

```python
toss_decision_prediction = log.predict(np.array(x_test).reshape(-1, 1))
toss_decision_prediction
```

```python
log.score(np.array(toss_decision_prediction).reshape(-1, 1),y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,toss_decision_prediction)
```

## How many no balls will be scored?

```python
RCB = deliveries.loc[(deliveries['bowling_team'] == 'Royal Challengers Bangalore')]
RCB
```

```python
KKR = deliveries.loc[(deliveries['bowling_team'] == 'Kolkata Knight Riders')]
KKR
```

```python
KKR = KKR.groupby('match_id')['noball_runs'].sum()
KKR
```

```python
Total_NOBALLS_by_RCB = pd.DataFrame({'total NOBALLS per match':RCB}) 

Total_NOBALLS_by_RCB
```

```python
Total_NOBALLS_by_KKR['total NOBALLS per match'].value_counts()
```

```python
Total_NOBALLS_by_RCB.describe()
```

```python
Total_NOBALLS_by_KKR.describe()
```

```python
fig, axes = plt.subplots(1, 2,  sharex=True, figsize=(10,5))

ax1 = sns.countplot(ax=axes[0],x='total NOBALLS per match',data=Total_NOBALLS_by_RCB,order = Total_NOBALLS_by_RCB['total NOBALLS per match'].value_counts().index)
ax1.set_title('visualization based on NOBALLS in each match by RCB')
ax2 = sns.countplot(ax=axes[1],x='total NOBALLS per match',data=Total_NOBALLS_by_KKR,order = Total_NOBALLS_by_KKR['total NOBALLS per match'].value_counts().index)
ax2.set_title('visualization based on NOBALLS in each match by KKR')
plt.show()
```
